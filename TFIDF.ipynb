```
import string
from string import digits
from sklearn.feature_extraction.text import TfidfVectorizer

import re

def token(text):
    text = text.split(' ')
    table = str.maketrans('', '', string.punctuation)
    table1 = str.maketrans('','', digits)
    reg = re.compile('[^a-zA-Z ]')
    stripped = [w.translate(table) for w in text]
    stripped1 = [w.translate(table1) for w in stripped]
    stripped2 = [reg.sub('', w) for w in stripped1]
    stripped3 = [w for w in stripped2 if len(w)>1]
    return stripped3
```

```
vect_Title = TfidfVectorizer( max_features= 200, tokenizer=token , stop_words = 'english').fit(df["Review Title"].values.astype('U'))

vect_Title_Train = vect_Title.transform(df["Review Title"])
vect_Title_Test = vect_Title.transform(df_test["Review Title"])

vect_Title_Train=vect_Title_Train.toarray()
vect_Title_Test=vect_Title_Test.toarray()

words = vect_Title.get_feature_names()
words=["T_"+w for w in words]

keywords_Train=pd.DataFrame(vect_Title_Train, columns=words)
keywords_Test=pd.DataFrame(vect_Title_Test, columns=words)

keys_Train=[w for w in words if keywords_Train[w].sum()>60] 
#keys_Test=[w for w in words if keywords_Test[w].sum()] 

col_train=pd.concat([df, keywords_Train[keys_Train]],axis=1)
col_test=pd.concat([df_test, keywords_Test[keys_Train]],axis=1)
```

```
vect_Title = TfidfVectorizer( max_features= 200, tokenizer=token , stop_words = 'english').fit(df["Review Text"].values.astype('U'))

vect_Title_Train = vect_Title.transform(df["Review Text"])
vect_Title_Test = vect_Title.transform(df_test["Review Text"])

vect_Title_Train=vect_Title_Train.toarray()
vect_Title_Test=vect_Title_Test.toarray()

words = vect_Title.get_feature_names()
words=["R_"+w for w in words]

keywords_Train=pd.DataFrame(vect_Title_Train, columns=words)
keywords_Test=pd.DataFrame(vect_Title_Test, columns=words)

keys_Train=[w for w in words if keywords_Train[w].sum()>60] 
#keys_Test=[w for w in words if keywords_Test[w].sum()] 

col_train=pd.concat([col_train, keywords_Train[keys_Train]],axis=1)
col_test=pd.concat([col_test, keywords_Test[keys_Train]],axis=1)
```

```
best = pd.DataFrame(best_t)
df_m = pd.concat([df_test, best], axis =1)
df_m.columns = ['Review Text','Review Title', 'new', 'wins', 'F', 'E', 'D', 'C', 'B', 'A']
df_m.head()
```

```
def fun(win, A, B, C, D, E, F): 
  
    if win == 1:
        return A
  
    elif win == 2: 
        return B
    
    elif win == 3: 
        return C
      
    elif win == 4: 
        return D
      
    elif win == 5: 
        return E
  
    else: 
        return F
        
```

```
df_m['try'] = df_m.apply(lambda x: fun(x['wins'], x['A'], x['B'], x['C'], x['D'], x['E'], x['F']), axis = 1)
```

```
df_s = df_m.drop([ 'new', 'wins', 'A', 'B', 'C', 'D', 'E', 'F'], axis  =1)
```

```
df_s['topic'] = label_encoder.inverse_transform(df_s['try'])
df_s = df_s.drop('try', axis = 1)
```
